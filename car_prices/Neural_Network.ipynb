```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.base import BaseEstimator
import numpy as np
```

```python
# Train ,test 
x = model_df.drop(columns=["sellingprice"])
y = model_df["sellingprice"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
training_score = []
testing_score = []
```

```python
# Define the custom wrapper class
class KerasRegressorWrapper(BaseEstimator):
    def __init__(self, optimizer='adam', activation='relu', neurons=100, epochs=1, batch_size=2048, verbose=0):
        self.optimizer = optimizer
        self.activation = activation
        self.neurons = neurons
        self.epochs = epochs
        self.batch_size = batch_size
        self.verbose = verbose
        self.model = self.build_model()

    def build_model(self):
        model = Sequential()
        model.add(Dense(self.neurons, input_dim=x_train.shape[1], activation=self.activation))
        model.add(Dense(1, activation='linear'))
        model.compile(optimizer=self.optimizer, loss='mean_squared_error')
        return model

    def fit(self, X, y):
        self.model.fit(X, y, epochs=self.epochs, batch_size=self.batch_size, verbose=self.verbose)
        return self

    def predict(self, X):
        return self.model.predict(X)
```

```python
# Create an instance of the custom wrapper
nn_model = KerasRegressorWrapper()

# Define the hyperparameter grid
param_grid = {
    'optimizer': ['adam', 'rmsprop'],
    'activation': ['relu', 'tanh'],
    'neurons': [50, 100, 150]
}
```

```python
# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=nn_model, param_grid=param_grid,
                           scoring=make_scorer(mean_squared_error, greater_is_better=False),
                           cv=3, n_jobs=-1)
```

```python
# Fit the grid search to the data
grid_search.fit(x_train, y_train)

# Print the best parameters
print("Best hyperparameters:", grid_search.best_params_)
```

```python
# Get the best model
best_nn_model = grid_search.best_estimator_

# Evaluate the best model
y_pred_train = best_nn_model.predict(x_train)
y_pred_test = best_nn_model.predict(x_test)

# Assess the performance
r2_train = r2_score(y_train, y_pred_train)
r2_test = r2_score(y_test, y_pred_test)
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))

print("R2 Score on training data:", r2_train)
print("R2 Score on testing data:", r2_test)
print("RMSE on training data:", rmse_train)
print("RMSE on testing data:", rmse_test)
```
